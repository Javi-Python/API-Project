```python
import os
import dotenv
from pymongo import MongoClient
```

```python
dotenv.load_dotenv()

dburl = os.getenv('URL')

print(dburl)
if not dburl:
    raise ValueError('no tienes url de mongo')

client = MongoClient(dburl)
db = client.get_database()
collection = db['WSB_posts']
```

```python
from config.configuration import db, collection

def ticker_posts(ticker):
    '''
    Funcion que devuelve todos los posts que incluyen cierto ticker en el titulo.
    '''

    query = {'tickers': f"{ticker}"}
    ticker_posts = list(collection.find(query, {'_id': 0}))
    return ticker_posts
```

```python
def top_posts():
    count = collection.aggregate([
    { "$group": { "_id": "$tickers", "count": { "$sum": 1 } } }, 
    { "$project":{ "_id": 0, "name": "$_id", "count":"$count" } },
    { '$sort': { 'count': -1}}])
    return count
```

```python
def title_sentiment_over(number):
    query = {'title_sentiment' : {'$gt' : number}} 
    t_sent = list(collection.find(query, {'_id':0}))
    return t_sent
```

```python
def title_sentiment_less(number):
    query = {'title_sentiment' : {'$lt' : number}} 
    t_sent = list(collection.find(query, {'_id':0}))
    return t_sent
```
